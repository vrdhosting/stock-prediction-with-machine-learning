## SubPlots and Legends

```python
days = [1,2,3,4,5];
original_price = [10,20,25,30,40];
predicted_price = [8,21,25,28,33];
```

```python
plt.plot(days, original_price, 'b-o', label='Original price');
plt.plot(days, predicted_price, 'r-o', label='Predicted price');
plt.title("Stock Price");
plt.xlabel('Days');
plt.ylabel('Price');
plt.legend();
```

```python
# First Subplot - Original Price
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.legend();
plt.title('Days Vs Original Price');
plt.xlabel('Days');
plt.ylabel('Original Price');

# Second Subplot - Pridicted Price
plt.subplot(3,1,2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend();
plt.title('Days Vs Predicted Price');
plt.xlabel('Days');
plt.ylabel('Predicted Price');


# Third Subplot - Difference Between Original and Predicted Prices
price_difference = [o-p for o,p in zip(original_price, predicted_price)];
plt.subplot(3,1,3)
plt.bar(days, price_difference, color = 'purple', label='Price Difference');
plt.legend();
plt.title('Days Vs Price Difference');
plt.xlabel('Days');
plt.ylabel('Price Difference');
plt.tight_layout();
# Save the Image as .png File
# plt.savefig('example_prediction.png', dpi=300) 
```
